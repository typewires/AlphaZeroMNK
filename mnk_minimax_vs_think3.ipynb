# Minimax vs. Think3: 6×6, 4-in-a-row game

**6,6,4-game (a special case of the more general m,n,k game):** a 6×6 board where **k = 4** in a row in any direction wins. X moves first (+1), O second (−1).

**Minimax (depth-limited, radiuzs-1):** in a standard minimax search, every empty cell may be considered at each move. The radius-1 restriction limits candidate moves to empty cells within Chebyshev distance <= 1 of an existing stone. The term radius-1 is used informally here for this neighborhood-based move generation rule. Restricting moves this way reduces the branching factor by focusing the search on locally relevant cells.

**Think3:** a simple 3-move lookahead heuristic.

## 100 games (headless, parallel games)

Minimax (depth=3, radius 1) vs. Think3. Alternate who plays first.

```python
results = run_match(100)
```

```python
from collections import Counter

c = Counter(results)
wins, losses, ties = c.get(1, 0), c.get(-1, 0), c.get(0, 0)
n = len(results)

first  = [results[i] for i in range(n) if i % 2 == 0]   # minimax moved first
second = [results[i] for i in range(n) if i % 2 == 1]    # minimax moved second
def wlt(rs): return rs.count(1), rs.count(-1), rs.count(0)
fw, fl, ft = wlt(first)
sw, sl, st = wlt(second)

print(f"Minimax vs Think3  —  {n} games\n")
print(f"  Wins   : {wins:3d}  ({wins/n:6.1%})")
print(f"  Losses : {losses:3d}  ({losses/n:6.1%})")
print(f"  Ties   : {ties:3d}  ({ties/n:6.1%})")
print(f"  Score  : {(wins - losses)/n:+.3f}  per game\n")
print(f"  Minimax first  (X): {fw}W / {fl}L / {ft}T")
print(f"  Minimax second (O): {sw}W / {sl}L / {st}T")
```
